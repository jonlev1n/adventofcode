```python
import copy
import time
import re
import math
# file = open("/Users/jonathanlevin/git/adventofcode2020/day17/input.txt", "r")
file = open("/Users/jonathanlevin/git/adventofcode2020/day17/test_input.txt", "r")
# file = open("/Users/jonathanlevin/git/adventofcode2020/day17/test_input2.txt", "r")

active_cubes = []
grid = []
lines = file.read().splitlines()
x = 0
for line in lines:
    y = 0
    for cube in line:
        coord = [x,y,0,0]
        grid.append(coord)
        if cube == '#':
            active_cubes.append(coord)
        
        y += 1
    x += 1

# print("t= 0, active: %s" % active_cubes)
# print(grid)
# if a coordinate is not in inital_active it is inactive

def part1(n):
    start_grid = copy.deepcopy(grid)
    for i in range(0, n):
        next_grid = []
        for cube in start_grid:
            for x in range(-1,2,1):
                for y in range(-1,2,1):
                    for z in range(-1,2,1):
                        if not [cube[0] + x, cube[1] + y, cube[2] + z] in next_grid:
                            next_grid.append(([cube[0] + x, cube[1] + y, cube[2] + z]))
        start_grid = copy.deepcopy(next_grid)
        cubes_to_add = []
        cubes_to_remove = []
        for cube in next_grid:
            neighbors = []
            for x in range(-1,2,1):
                for y in range(-1,2,1):
                    for z in range(-1,2,1):
                        if not (x == 0 and y == 0 and z == 0):
                            neighbors.append([cube[0] + x, cube[1] + y, cube[2] + z])

            active_neighbors = 0
            for neighbor in neighbors:
                if neighbor in active_cubes:
                    active_neighbors += 1

            if cube in active_cubes and active_neighbors not in [2,3]:
                if cube not in cubes_to_remove:
                    cubes_to_remove.append(cube)
            elif cube not in active_cubes and active_neighbors == 3:
                if cube not in cubes_to_add:
                    cubes_to_add.append(cube)

        for cube in cubes_to_add:
            active_cubes.append(cube)
        for cube in cubes_to_remove:
            active_cubes.remove(cube)

        # print("add: %s" % cubes_to_add)
        # print("remove: %s" % cubes_to_remove)
        # print("active: %s" % active_cubes)
    print("There are %d active cubes" % len(active_cubes))
    
def part2(n):
    start_grid = copy.deepcopy(grid)
    for i in range(0, n):
        next_grid = []
        for cube in start_grid:
            for x in range(-1,2,1):
                for y in range(-1,2,1):
                    for z in range(-1,2,1):
                        for w in range(-1,2,1):
                            if not [cube[0] + x, cube[1] + y, cube[2] + z, cube[3] + w] in next_grid:
                                next_grid.append(([cube[0] + x, cube[1] + y, cube[2] + z, cube[3] + w]))
        start_grid = copy.deepcopy(next_grid)
        cubes_to_add = []
        cubes_to_remove = []
        for cube in next_grid:
            neighbors = []
            for x in range(-1,2,1):
                for y in range(-1,2,1):
                    for z in range(-1,2,1):
                        for w in range(-1,2,1):
                            if not (x == 0 and y == 0 and z == 0 and w == 0):
                                neighbors.append([cube[0] + x, cube[1] + y, cube[2] + z, cube[3] + w])
            active_neighbors = 0
            for neighbor in neighbors:
                if neighbor in active_cubes:
                    active_neighbors += 1

            if cube in active_cubes and active_neighbors not in [2,3]:
                if cube not in cubes_to_remove:
                    cubes_to_remove.append(cube)
            elif cube not in active_cubes and active_neighbors == 3:
                if cube not in cubes_to_add:
                    cubes_to_add.append(cube)

        for cube in cubes_to_add:
            active_cubes.append(cube)
        for cube in cubes_to_remove:
            active_cubes.remove(cube)

        # print("add: %s" % cubes_to_add)
        # print("remove: %s" % cubes_to_remove)
        # print("active: %s" % active_cubes)
    print("There are %d active cubes" % len(active_cubes))
                

# part1(6)
part2(6)
```
